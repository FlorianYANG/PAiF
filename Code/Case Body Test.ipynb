```python
import pandas as pd
```

```python
df = pd.read_excel(r'D:\Programming\Python\bodytest.xlsx', index_col='Order')
df
```

```python
res = df.pivot_table(values='Result', index=['UserId','TestId'], columns='TestName', aggfunc='last')
res
```

```python
res = res.reset_index()
res['Year'] = res['TestId'].astype(str).str[:2]
res
```

```python
uni_id = res.UserId.unique().tolist()
tmp_dict = {i:{} for i in uni_id}
for i in uni_id:
    for col in ['GGT', 'WBC', 'GLU']:
        for year in ['19', '20', '21']:
            try:
                tmp_dict[i][col+year] = res[(res['UserId']==i) & (res['Year']==year)][col].values[0]
            except:
                pass
```

```python
res2 = pd.DataFrame.from_dict(tmp_dict, orient='index')
res2
```

```python
info = df[['Name', 'UserId']].drop_duplicates()
info.merge(res2, left_on='UserId', right_index=True)
```

## better way to do

```python
df2 = df.copy()
df2['Year'] = df.TestId.astype(str).str[:2]
df2['TestName'] = df2['TestName'] + df2['Year']
df2 = df2.sort_index()
df2
```

```python
df2 = df2[['UserId', 'TestName', 'Result']]
df2 = df2.set_index(['UserId', 'TestName']).unstack()
df2
```

```python
info = df[['Name', 'UserId']].drop_duplicates()
info.merge(df2, left_on='UserId', right_index=True)
```
